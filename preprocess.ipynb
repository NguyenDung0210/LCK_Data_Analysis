```python
import pandas as pd
```

```python
df = pd.read_csv('LCK_summer_2024.csv')
df
```

```python
print(df['KDA'].dtype)
```

```python
kda_col = None
for col in df.columns:
    if "KDA" in col:
        kda_col = col
        break

mask = df[kda_col] == "Perfect KDA"
df.loc[mask, kda_col] = df.loc[mask].apply(
    lambda row: int(row['Kills']) + int(row['Assists']), axis=1
)

df
```

```python
print(df['KDA'].dtype)
```

```python
df['KDA'] = df['KDA'].astype(float)
df['KDA'].dtype
```
