```python
def get_game_ids_from_tournament(tournament_url):
    response = requests.get(tournament_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    links = soup.find_all('a', href=True)
    game_ids = []

    for link in links:
        href = link['href']
        match = re.search(r'/game/stats/(\d+)/', href)
        if match:
            game_id = int(match.group(1))
            game_ids.append(game_id)

    return list(set(game_ids))

tournament_url = "https://gol.gg/tournament/tournament-matchlist/LCK%20Summer%202024/"
game_ids = get_game_ids_from_tournament(tournament_url)

print(f"Found {len(game_ids)} games")
print(sorted(game_ids))
```

```python
def scrape_game(game_id):
    url = f"https://gol.gg/game/stats/{game_id}/page-fullstats/"
    options = Options()
    options.use_chromium = True
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    service = Service(EdgeChromiumDriverManager().install())
    driver = webdriver.Edge(service=service, options=options)
    
    driver.get(url)
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()
    
    table = soup.find("table", class_="completestats")
    if not table:
        print(f"Game {game_id} has no data.")
        return None

    thead = table.find("thead")
    champion_names = [img['alt'] for img in thead.find_all("img")]

    tbody = table.find("tbody")
    rows = tbody.find_all("tr")
    data_rows = [[col.get_text(strip=True) for col in row.find_all("td")] for row in rows]
    
    df = pd.DataFrame(data_rows)
    if df.empty: 
        return None

    labels = df[0].tolist()
    data = df.iloc[:, 1:].T
    data.columns = labels
    data.insert(0, "Champion", champion_names)
    data.insert(0, "GameID", game_id)
    return data

all_data = []
for i, gid in enumerate(game_ids):
    print(f"Game {i+1}/{len(game_ids)} - ID {gid}")
    df = scrape_game(gid)
    if df is not None:
        all_data.append(df)
    time.sleep(1)

df_all = pd.concat(all_data, ignore_index=True)
df_all.to_csv("LCK_all_games.csv", index=False)
df_all.head()
```
