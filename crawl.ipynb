```python
def get_match_ids_from_tournament(tournament_url):
    response = requests.get(tournament_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    links = soup.find_all('a', href=True)
    match_ids = []

    for link in links:
        href = link['href']
        match = re.search(r'/game/stats/(\d+)/', href)
        if match:
            match_id = int(match.group(1))
            match_ids.append(match_id)

    return list(set(match_ids))

tournament_url = "https://gol.gg/tournament/tournament-matchlist/LCK%20Summer%202024/"
match_ids = sorted(get_match_ids_from_tournament(tournament_url))

print(f"Found {len(match_ids)} matches")
print(match_ids)
```

```python
def scrape_game(game_id):
    options = Options()
    options.use_chromium = True
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    url_stats = f"https://gol.gg/game/stats/{game_id}/page-fullstats/"
    driver = webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()), options=options)
    driver.get(url_stats)
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()
    
    table = soup.find("table", class_="completestats")
    if not table:
        print(f"{game_id} has no data.")
        return None

    thead = table.find("thead")
    champion_names = [img['alt'] for img in thead.find_all("img")]

    tbody = table.find("tbody")
    rows = tbody.find_all("tr")
    data_rows = [[col.get_text(strip=True) for col in row.find_all("td")] for row in rows]
    
    df = pd.DataFrame(data_rows)
    if df.empty: 
        return None

    labels = df[0].tolist()
    data = df.iloc[:, 1:].T
    data.columns = labels
    data.insert(1, "Champion", champion_names)

    url_game = f"https://gol.gg/game/stats/{game_id}/page-game/"
    driver = webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()), options=options)
    driver.get(url_game)
    time.sleep(2)
    soup_game = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()

    blue_team_div = soup_game.find("div", class_="col-4 blue-line-header text-center")
    red_team_div = soup_game.find("div", class_="col-4 red-line-header text-center")
    teams = [
        blue_team_div.get_text(strip=True),
        red_team_div.get_text(strip=True)
    ]

    result_divs = soup_game.find_all("div", class_="col-12 blue-line-header") + \
                soup_game.find_all("div", class_="col-12 red-line-header")
    results = []
    for div in result_divs[:2]:
        text = div.get_text(strip=True)
        if "WIN" in text:
            results.append("Win")
        elif "LOSS" in text:
            results.append("Loss")
        else:
            results.append("UNKNOWN")

    teams =  [teams[0]] * 5 + [teams[1]] * 5
    results = [results[0]] * 5 + [results[1]] * 5

    data.insert(0, "Team", teams)
    data.insert(1, "Result", results)

    data.insert(0, "GameID", game_id)
    return data

all_data = []
count = 1
for match_id in match_ids:
    for i in range(3):
        game_id = match_id + i
        print(f"Game {count} - ID {game_id}")
        df = scrape_game(game_id)
        if df is not None:
            all_data.append(df)
        count += 1
        time.sleep(1)

df_all = pd.concat(all_data, ignore_index=True)
df_all.to_csv("LCK_summer_2024.csv", index=False)
df_all.head()
```
